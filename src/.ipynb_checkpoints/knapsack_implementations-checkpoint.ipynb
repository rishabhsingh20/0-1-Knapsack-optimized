## Table Based Approach

```python
def knapSackTable(W, wt, val, n):
    start = timer()
    DP = [[0 for x in range(W + 1)] for x in range(n + 1)]   # DP Matrix consisting of W columns and n rows
 
    # Build table DP[][] in bottom up manner
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:          # If no.of items= 1   or  Capacity of Knapsack= 0(column 1),
                DP[i][w] = 0              # then it will have value 0
            elif wt[i-1] <= w:            
                DP[i][w] = max(val[i-1]   
                          + DP[i-1][w-wt[i-1]],      # Filling values according to the equation mentioned in the paper
                              DP[i-1][w])
            else:
                DP[i][w] = DP[i-1][w]   
    end = timer()
    time = end-start
    return DP[n][W],time
```

## List Based Approach

```python
def merge(ls1, ls2):                        # Merging the two list of pairs
    ls = []
    i = 0
    j = 0
    l1 = len(ls1)
    l2 = len(ls2)
    val_temp = []
    w_temp = []
    while(i<l1 and j<l2):
        val1,w1 = ls1[i]
        val2,w2 = ls2[j]
        if w1<=w2 and val1<=val2:         # Only keep the dominant intervals
            i+=1
        elif w2<=w1 and val2 <= val1:
            j+=1
        elif w1>=w2:                      # Storing dominant pairs only into new list
            val_temp.append(val1)
            w_temp.append(w1)
            i+=1
        else:
            val_temp.append(val2)
            w_temp.append(w2)
            j+=1
    while(i<l1):
        val1,w1 = ls1[i]
        val_temp.append(val1)
        w_temp.append(w1)
        i+=1
    while(j<l2):
        val2,w2 = ls2[j]
        val_temp.append(val2)
        w_temp.append(w2)
        j+=1
    ls = list(zip(val_temp,w_temp))
    return ls

def knapsackList(W, wt, val, n):
    start = timer()
    ls = [(0,W)]            # List of pairs. The first entry will be (0,C) or here it is (0,W)
    for i in range(n):
        val_temp = []
        w_temp = []
        for j in range(len(ls)):
            v,w = ls[j]
            if w - wt[i] >= 0:                 # Checking for dominant pairs
                val_temp.append(v + val[i])    # and putting them into another "list of pairs" containing the updated
                w_temp.append(w - wt[i])       # profits and weights for the current iteration
        ls_temp = list(zip(val_temp,w_temp))   # Merging this new list with the original 
        ls = merge(ls_temp,ls)                 # and updating the original

    val,w = ls[len(ls)-1]
    end = timer()
    time = end - start
    return val,time
```

## List Based Approach Accelerated using UB1 (Upper Bound -1)

```python
def ubound1(val,w,val_v,wt_v):                # Calculating Upper Bound 1
    return val + (val_v/wt_v)*w

def merge_ub1(ls1, ls2, maxi, val, wt):        # Similar merge function as in the list-based appraoch, but
    ls = []                                    # catering for the upper bound
    i = 0
    j = 0
    l1 = len(ls1)
    l2 = len(ls2)
    val_temp = []
    w_temp = []
    while(i<l1 and j<l2):
        val1,w1 = ls1[i]
        val2,w2 = ls2[j]
        if ubound1(val1,w1,val,wt) < maxi:     # We calculate the upper bound and if it is greater than the current 
            i+=1                               # maximum profit (maxi) then we neglect it
            continue
        elif ubound1(val2,w2,val,wt) < maxi:
            j+=1
            continue
        elif w1<=w2 and val1<=val2:
            i+=1
        elif w2<=w1 and val2 <= val1:
            j+=1
        elif w1>=w2:
            val_temp.append(val1)
            w_temp.append(w1)
            i+=1
        else:
            val_temp.append(val2)
            w_temp.append(w2)
            j+=1
    while(i<l1):
        val1,w1 = ls1[i]
        val_temp.append(val1)
        w_temp.append(w1)
        i+=1
    while(j<l2):
        val2,w2 = ls2[j]
        val_temp.append(val2)
        w_temp.append(w2)
        j+=1
    ls = list(zip(val_temp,w_temp))
    return ls

def knapsack_ub1(W, wt, val, n):              # Similar to knapsackList function seen earlier. 
    start = timer()
    ls = [(0,W)]
    for i in range(n):
        val_temp = []
        w_temp = []
        for j in range(len(ls)):
            v,w = ls[j]
            if w - wt[i] >= 0:
                val_temp.append(v + val[i])
                w_temp.append(w - wt[i])
        if len(val_temp) > 0:
            maxi = max(val_temp)
            ls_temp = list(zip(val_temp,w_temp))
            ls = merge_ub1(ls_temp,ls,maxi,val[i],wt[i])

    val,w = ls[len(ls)-1]
    end = timer()
    time = end - start
    return val,time
```

## List Based Appraoch Accelerated using UB2 (Upper Bound - 2)

```python
def ubound2(val,w,k,val_v,wt_v):             # Calculating Upper Bound 2 as described in the paper
    bound = 0.0
    while k < len(wt_v)-1 and w >= wt_v[k]:
        bound = bound + val_v[k];
        w = w- wt_v[k]
        k+=1
    bound += val + (val_v[k]/wt_v[k])*w
    return bound
        
def merge_ub2(ls1, ls2, maxi, k, val, wt):     # The merge operations are same except for the fact that 
    ls = []                                    # we compare the maximum current profit(maxi) with the new Upper Bound 2
    i = 0
    j = 0
    l1 = len(ls1)
    l2 = len(ls2)
    val_temp = []
    w_temp = []
    while(i<l1 and j<l2):
        val1,w1 = ls1[i]
        val2,w2 = ls2[j]
        if ubound2(val1,w1,k, val,wt) < maxi:
            i+=1
            continue
        elif ubound2(val2,w2,k, val,wt) < maxi:
            j+=1
            continue
        elif w1<=w2 and val1<=val2:
            i+=1
        elif w2<=w1 and val2 <= val1:
            j+=1
        elif w1>=w2:
            val_temp.append(val1)
            w_temp.append(w1)
            i+=1
        else:
            val_temp.append(val2)
            w_temp.append(w2)
            j+=1
    while(i<l1):
        val1,w1 = ls1[i]
        val_temp.append(val1)
        w_temp.append(w1)
        i+=1
    while(j<l2):
        val2,w2 = ls2[j]
        val_temp.append(val2)
        w_temp.append(w2)
        j+=1
    ls = list(zip(val_temp,w_temp))
    return ls

def knapsack_ub2(W, wt, val, n):                     # Same as the knapsack_ub1 function. 
    start = timer()                                 

    ls = [(0,W)]
    for i in range(n):
        val_temp = []
        w_temp = []
        for j in range(len(ls)):
            v,w = ls[j]
            if w - wt[i] >= 0:
                val_temp.append(v + val[i])
                w_temp.append(w - wt[i])
        if len(val_temp) > 0:
            maxi = max(val_temp)
            ls_temp = list(zip(val_temp,w_temp))
            ls = merge_ub2(ls_temp,ls,maxi,i,val,wt)

    val,w = ls[len(ls)-1]
    
    end = timer()
    time = end - start 
    
    return val,time
```

## Function for generating testcases

```python
def Sort_Tuple(tup): 
    tup.sort(key = lambda x: x[0]) 
    return tup 

def generate_testcase(n,alpha):     
    val = random.sample(range(1, n*20), n)
    wt = random.sample(range(20, n*10), n)
        
    efficiency = [i / j for i, j in zip(val, wt)]    
    lst_tuple = list(zip(efficiency,val,wt))
    sortedT = Sort_Tuple(lst_tuple)
        
    for j in range(len(sortedT)):
        val[n-1-j] = sortedT[j][1]
        wt[n-1-j] = sortedT[j][2]
            
    W = int(sum(wt) * alpha)
        
    return val,wt,W
```

## Function for getting Comparison Plot of the Different Implementations

```python
def plotResult(high,alpha):  
    Y1 = []
    Y2 = []
    Y3= []
    Y4 = []
    X = []
    filename = './result_for_alpha='+str(alpha)+'.png'
    
    for n in range(5,high,5):
        X.append(n)
        val, wt, W = generate_testcase(n,alpha)
        KST, t1 = knapSackTable(W, wt, val, n)
        KSL, t2 = knapsackList(W, wt, val, n)
        KSUB1, t3 = knapsack_ub1(W, wt, val, n)
        KSUB2 , t4 = knapsack_ub2(W, wt, val, n)
        
        Y1.append(t1)
        Y2.append(t2)
        Y3.append(t3)
        Y4.append(t4)
        
    plt.figure(figsize=(6,6))
    plt.plot(X, Y1)
    plt.plot(X, Y2)
    plt.plot(X, Y3)
    plt.plot(X, Y4)
    plt.xlabel("Number of Terms")
    plt.ylabel("Time taken")
    plt.legend(['Tabular','List based','UB1','UB2'])
    plt.title("Comparison of Runtimes between 4 Implementations of Knapsack Problem")
    plt.savefig(filename)

    return X,Y1,Y2,Y3,Y4
```

```python
def plotResults(high,alphas):  
    Y1 = []
    Y2 = []
    Y3= []
    Y4 = []
    X = [i for i in range(5,high,5)]
    
    for i in alphas:
        y1,y2,y3,y4 = [],[],[],[]
        for n in range(5,high,5):
            val, wt, W = generate_testcase(n,i)
            KST, t1 = knapSackTable(W, wt, val, n)
            KSL, t2 = knapsackList(W, wt, val, n)
            KSUB1, t3 = knapsack_ub1(W, wt, val, n)
            KSUB2 , t4 = knapsack_ub2(W, wt, val, n)

            y1.append(t1)
            y2.append(t2)
            y3.append(t3)
            y4.append(t4)
        Y1.append(y1)
        Y2.append(y2)
        Y3.append(y3)
        Y4.append(y4)
        
    fig,axs = plt.subplots(len(alphas)//2,len(alphas)//2,figsize=(12,12))
    idx = 0
    for i in range(len(alphas)//2):
        for j in range(len(alphas)//2):
            axs[i,j].plot(X, Y1[idx])
            axs[i,j].plot(X, Y2[idx])
            axs[i,j].plot(X, Y3[idx])
            axs[i,j].plot(X, Y4[idx])
            
            axs[i,j].set_xlabel("Number of Terms")
            axs[i,j].set_ylabel("Time taken")
            axs[i,j].legend(['Tabular','List based','UB1','UB2'])
            axs[i,j].set_title('Comparing Runtime for alpha = '+str(alphas[idx]))
            idx += 1      
    fig.suptitle('Runtime Comparison for the 4 different approaches for different values of Alpha')
    plt.savefig('result_for_multipe_alphas.png')
    plt.show()
```
