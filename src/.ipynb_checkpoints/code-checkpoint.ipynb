#### Four types of implementations
- Table based (conventional Dynamic programming approach)
- List based
- List based accelerated using UB1
- List based accelerated using UB2

#### The function below is the Table Based Approach

```python
def knapSackTable(W, wt, val, n):
    start = timer()
    DP = [[0 for x in range(W + 1)] for x in range(n + 1)]
 
    # Build table DP[][] in bottom up manner
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                DP[i][w] = 0
            elif wt[i-1] <= w:
                DP[i][w] = max(val[i-1]
                          + DP[i-1][w-wt[i-1]], 
                              DP[i-1][w])
            else:
                DP[i][w] = DP[i-1][w]
    end = timer()
    time = end-start
    return DP[n][W],time
```

#### The functions given below is for List Based Approach

```python
def merge(ls1, ls2):
    ls = []
    i = 0
    j = 0
    l1 = len(ls1)
    l2 = len(ls2)
    val_temp = []
    w_temp = []
    while(i<l1 and j<l2):
        val1,w1 = ls1[i]
        val2,w2 = ls2[j]
        if w1<=w2 and val1<=val2:
            i+=1
        elif w2<=w1 and val2 <= val1:
            j+=1
        elif w1>=w2:
            val_temp.append(val1)
            w_temp.append(w1)
            i+=1
        else:
            val_temp.append(val2)
            w_temp.append(w2)
            j+=1
    while(i<l1):
        val1,w1 = ls1[i]
        val_temp.append(val1)
        w_temp.append(w1)
        i+=1
    while(j<l2):
        val2,w2 = ls2[j]
        val_temp.append(val2)
        w_temp.append(w2)
        j+=1
    ls = list(zip(val_temp,w_temp))
    return ls

def knapsackList(W, wt, val, n):
    start = timer()
    ls = [(0,W)]
    for i in range(n):
        val_temp = []
        w_temp = []
        for j in range(len(ls)):
            v,w = ls[j]
            if w - wt[i] >= 0:
                val_temp.append(v + val[i])
                w_temp.append(w - wt[i])
        ls_temp = list(zip(val_temp,w_temp))
        ls = merge(ls_temp,ls)

    val,w = ls[len(ls)-1]
    end = timer()
    time = end - start
    return val,time
```

#### The functions given below are for Knapsack accelerated using UB1 (Upper Bound -1)

```python
def ubound1(val,w,val_v,wt_v):
    return val + (val_v/wt_v)*w

def merge_ub1(ls1, ls2, maxi, val, wt):
    ls = []
    i = 0
    j = 0
    l1 = len(ls1)
    l2 = len(ls2)
    val_temp = []
    w_temp = []
    while(i<l1 and j<l2):
        val1,w1 = ls1[i]
        val2,w2 = ls2[j]
        if ubound1(val1,w1,val,wt) < maxi:
            i+=1
            continue
        elif ubound1(val2,w2,val,wt) < maxi:
            j+=1
            continue
        elif w1<=w2 and val1<=val2:
            i+=1
        elif w2<=w1 and val2 <= val1:
            j+=1
        elif w1>=w2:
            val_temp.append(val1)
            w_temp.append(w1)
            i+=1
        else:
            val_temp.append(val2)
            w_temp.append(w2)
            j+=1
    while(i<l1):
        val1,w1 = ls1[i]
        val_temp.append(val1)
        w_temp.append(w1)
        i+=1
    while(j<l2):
        val2,w2 = ls2[j]
        val_temp.append(val2)
        w_temp.append(w2)
        j+=1
    ls = list(zip(val_temp,w_temp))
    return ls

def knapsack_ub1(W, wt, val, n):
    start = timer()
    ls = [(0,W)]
    for i in range(n):
        val_temp = []
        w_temp = []
        for j in range(len(ls)):
            v,w = ls[j]
            if w - wt[i] >= 0:
                val_temp.append(v + val[i])
                w_temp.append(w - wt[i])
        if len(val_temp) > 0:
            maxi = max(val_temp)
            ls_temp = list(zip(val_temp,w_temp))
            ls = merge_ub1(ls_temp,ls,maxi,val[i],wt[i])

    val,w = ls[len(ls)-1]
    end = timer()
    time = end - start
    return val,time
```

#### The functions given below are for Knapsack accelerated using UB2 (Upper Bound - 2)

```python
def ubound2(val,w,k,val_v,wt_v):
    bound = 0.0
    while k < len(wt_v)-1 and w >= wt_v[k]:
        bound = bound + val_v[k];
        w = w- wt_v[k]
        k+=1
    bound += val + (val_v[k]/wt_v[k])*w
    return bound
        
def merge_ub2(ls1, ls2, maxi, k, val, wt):
    ls = []
    i = 0
    j = 0
    l1 = len(ls1)
    l2 = len(ls2)
    val_temp = []
    w_temp = []
    while(i<l1 and j<l2):
        val1,w1 = ls1[i]
        val2,w2 = ls2[j]
        if ubound2(val1,w1,k, val,wt) < maxi:
            i+=1
            continue
        elif ubound2(val2,w2,k, val,wt) < maxi:
            j+=1
            continue
        elif w1<=w2 and val1<=val2:
            i+=1
        elif w2<=w1 and val2 <= val1:
            j+=1
        elif w1>=w2:
            val_temp.append(val1)
            w_temp.append(w1)
            i+=1
        else:
            val_temp.append(val2)
            w_temp.append(w2)
            j+=1
    while(i<l1):
        val1,w1 = ls1[i]
        val_temp.append(val1)
        w_temp.append(w1)
        i+=1
    while(j<l2):
        val2,w2 = ls2[j]
        val_temp.append(val2)
        w_temp.append(w2)
        j+=1
    ls = list(zip(val_temp,w_temp))
    return ls

def knapsack_ub2(W, wt, val, n):
    start = timer()

    ls = [(0,W)]
    for i in range(n):
        val_temp = []
        w_temp = []
        for j in range(len(ls)):
            v,w = ls[j]
            if w - wt[i] >= 0:
                val_temp.append(v + val[i])
                w_temp.append(w - wt[i])
        if len(val_temp) > 0:
            maxi = max(val_temp)
            ls_temp = list(zip(val_temp,w_temp))
            ls = merge_ub2(ls_temp,ls,maxi,i,val,wt)

    val,w = ls[len(ls)-1]
    
    end = timer()
    time = end - start 
    
    return val,time
```

```python
def Sort_Tuple(tup): 
    tup.sort(key = lambda x: x[0]) 
    return tup 

def generate_testcase(n,alpha):
    val = random.sample(range(1, n*20), n)
    wt = random.sample(range(20, n*10), n)
        
    efficiency = [i / j for i, j in zip(val, wt)]    
    lst_tuple = list(zip(efficiency,val,wt))
    sortedT = Sort_Tuple(lst_tuple)
        
    for j in range(len(sortedT)):
        val[n-1-j] = sortedT[j][1]
        wt[n-1-j] = sortedT[j][2]
            
    W = int(sum(wt) * alpha)
        
    return val,wt,W
```

#### function to plot the desired graph

```python
def plotResult():  
    Y1 = []
    Y2 = []
    Y3= []
    Y4 = []
    X = []
    
    for n in range(5,50,5):
        X.append(n)
        val, wt, W = generate_testcase(n,0.3)
        KST, t1 = knapSackTable(W, wt, val, n)
        KSL, t2 = knapsackList(W, wt, val, n)
        KSUB1, t3 = knapsack_ub1(W, wt, val, n)
        KSUB2 , t4 = knapsack_ub2(W, wt, val, n)
        
        Y1.append(t1)
        Y2.append(t2)
        Y3.append(t3)
        Y4.append(t4)
        
    plt.plot(X, Y1)
    plt.plot(X, Y2)
    plt.plot(X, Y3)
    plt.plot(X, Y4)
    plt.savefig('../Result/result.png')
```
